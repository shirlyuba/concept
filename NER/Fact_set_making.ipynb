```python
import os
import numpy as np
```

```python
path = r'devset/'
```

```python
list_of_txt = {}
list_of_spans = {}
list_of_objects = {}
tokens={}
for (dirpath, dirnames, filenames) in os.walk(path):
    for filename in filenames:
        if filename.endswith('.tokens'): 
            list_of_txt[filename] = os.sep.join([dirpath, filename])
        if filename.endswith('.spans'):
            list_of_spans[filename] = os.sep.join([dirpath, filename])
        if filename.endswith('.objects'):
            list_of_objects[filename] = os.sep.join([dirpath, filename])
```

```python
list_of_txt
```

```python
data_token_long = {}
i = 1
end_ind = []
for ind in list_of_spans.keys():
    f_span = open(list_of_spans[ind], 'r', encoding='utf8')
    name = ind.split('.')[0]
    data_token_long[name] = []
    pred = ""
    for line in f_span:
        array = line.split()
        if 'descr' in array[1] or array[1]=='geo_adj' or array[1]=='patronymic' or  array[1]=='prj_name':
            continue
        print(array)
        for j in range(int(array[5])):
            if j == 0 and array[1] == 'surname' and pred == 'name':
                end_ind.append(array[-(int(array[5]) * 2)])
            if j != 0:
                end_ind.append(array[-(int(array[5]) * 2 - j)])
            data_token_long[name].append([array[-(int(array[5]) * 2 - j) ], array[-(int(array[5])-j)], array[1]])
            if array[1] not in tokens.keys():
                tokens[array[1]] = i
                i += 1
        pred = array[1]
```

```python
tokens = {'geo_adj': 0,
 'job': 4,
 'loc_descr': 1,
 'loc_name': 1,
 'name': 2,
 'nickname': 2,
 'org_descr': 3,
 'org_name': 3,
 'patronymic': 0,
 'prj_descr': 0,
 'prj_name': 0,
 'surname': 2}
```

```python
tokens
```

```python
txt_ind = []
data_txt = []
for ind in list_of_txt.keys():
    f_txt = open(list_of_txt[ind], 'r', encoding='utf8')
    name = ind.split('.')[0]
    data_txt.append([])
    txt_ind.append([])
    for line in f_txt:
        if len(line) > 1:
            data_txt[-1].append(line.split()[-1])
            txt_ind[-1].append(line.split()[0])
```

```python
y = []
mask = []
for i in range(len(data_txt)):
    text = txt_ind[i]
    y.append(np.zeros(len(data_txt[i])))
    mask.append(np.zeros(len(data_txt[i])))
    ind = 0
    for tok in data_token_long[list(data_token_long.keys())[i]]:
        ind = text.index(tok[0])
        if tok[0] in end_ind:
            mask[-1][ind] = 1
        y[-1][ind]= tokens[tok[-1]]  
```

```python
NER = {
    0 : 'O',
    1 : '-LOC',
    2 : '-PER'
    3 : 'O',
    4 : 'O'
}
```

```python
new_y = []
for mas in y:
    mas = np.where(mas < 2, mas, 0)
    new_y.append(mas)
```

```python
with open('NER_Fact.txt', 'a', encoding='utf-8') as f:
    for i, (x1, y1) in enumerate(zip(data_txt, y)):
        f.writelines("-DOCSTART-\n")
        for j, (x2, y2) in enumerate(zip(x1, y1)):
            print()
            if y2 == 0:
                f.writelines(str(x2) + " " + NER[int(y2)] + '\n')
            else:
                if mask[i][j] == 1:
                    f.writelines(str(x2) + " I" + NER[int(y2)] + '\n')
                else:
                    f.writelines(str(x2) + " B" + NER[int(y2)] + '\n')
```
